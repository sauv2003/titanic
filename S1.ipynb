```python
import pandas as pd

data = {
    'Name': ['Amit', 'Ravi', 'Sneha', 'Priya', 'Rahul'],
    'Age': [25, 30, 22, 28, 24],
    'City': ['Delhi', 'Mumbai', 'Pune', 'Delhi', 'Bangalore'],
    'Marks': [85, 78, 92, 89, 75]
}

df = pd.DataFrame(data)
```

```python
print(df.loc[0])       # First row (by label)
print(df.iloc[0])      # First row (by position)
```

```python
print(df.loc[0:2])     # rows 0,1,2 (inclusive)
print(df.iloc[0:3])    # rows 0,1,2 (exclusive)
```

```python
print(df.loc[0:2, ['Name', 'Marks']])
```

```python
print(df.iloc[0:2, [0, 3]]  # first two rows, Name and Marks columns
)
```

```python
print(df[(df['City'] == 'Delhi') | (df['City'] == 'Pune')])
```

### Handling Missing Data

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Aarav", "Bhavna", "Chirag", "Diya", "Esha", "Farhan"],
    "Age": [25, 30, np.nan, 22, 28, np.nan],
    "City": ["Mumbai", "Delhi", None, "Kolkata", "Bangalore", "Pune"],
    "Score": [85, np.nan, 75, 90, np.nan, 88]
}

df = pd.DataFrame(data)
print(df)
```

```python
print(df.isnull())     # True where data is missing
print(df.isnull().sum())  # Total missing values per column
```

```python
df_drop_any = df.dropna()
```

```python
print(df_drop_any)
```

```python
df_drop_all = df.dropna(how="all")
print(df_drop_all)
```

```python
#Drop rows where specific columns have NaN:
df_drop_age = df.dropna(subset=["Age"])
```

```python
import pandas as pd

data = {
    "Name": ["Aarav", "Bhavna", "Chirag", "Diya", "Esha", "Farhan"],
    "Maths": [85, 90, 78, 92, 88, 76],
    "Science": [80, 95, 70, 89, 84, 82],
    "English": [75, 89, 68, 90, 85, 80],
    "City": ["Mumbai", "Delhi", "Pune", "Kolkata", "Delhi", "Pune"]
}

df = pd.DataFrame(data)
print(df)
```

```python
print(df.describe())       # Summary of numeric columns
print(df.mean(numeric_only=True))     # Column-wise mean
print(df.median(numeric_only=True))   # Median of each column
print(df.std(numeric_only=True))      # Standard deviation
print(df.var(numeric_only=True))      # Variance
```

```python
# Column-wise mean (default)
print(df.mean(numeric_only=True))

# Row-wise mean (mean marks per student)
df["Average"] = df[["Maths", "Science", "English"]].mean(axis=1)
print(df[["Name", "Average"]])
```

```python
# Sort by a column
df_sorted = df.sort_values(by="Science", ascending=False)

# Rank students based on Science marks
df["Science_Rank"] = df["Science"].rank(ascending=False)
print(df[["Name", "Science", "Science_Rank"]])
```

```python
# Group by City and find average marks
city_group = df.groupby("City")[["Maths", "Science", "English"]].mean()
print(city_group)

# Group by City and count how many students in each
print(df.groupby("City")["Name"].count())
```

```python

print(df.groupby("City").agg({
        "Maths": ["mean", "max", "min"],
        "Science": ["mean", "std"]
    })["Maths"]["mean"]
)
```

```python
import pandas as pd

# Create sample DataFrame
df = pd.DataFrame({
    "City": ["Pune", "Mumbai", "Pune", "Delhi", "Pune", "Mumbai", "Delhi", 
             "Chennai", "Chennai", "Bangalore", "Bangalore"],
    "Maths": [85, 92, 78, 88, 90, 85, 82, 95, 89, 87, 91],
    "Science": [90, 88, 92, 85, 87, 89, 80, 92, 88, 86, 90],
    "English": [88, 85, 90, 82, 89, 87, 78, 91, 86, 84, 89]
})

#1
# print(df[["Maths","Science","English"]].agg(["mean","median","std"]))

#2
df["Average"] = df[["Maths","Science","English"]].mean(axis = 1)
# print(df[["Name","Average"]])

#3
df["Total"] = df[["Maths","Science","English"]].sum(axis=1)
df["Rank"] = df["Total"].rank(ascending=False)
#4
df.sort_values(["English"],inplace=True)
df.sort_values(["Science"],ascending=False,inplace=True)
# print(df)
#5
print(df.groupby("City")["Total"].mean())

#6
city_maths_mean = df.groupby("City")["Maths"].transform('mean')
df["Flag"] = df["Maths"] > city_maths_mean
print(df)

#7
print(df.corr(numeric_only=True))

city_avg_totals = df.groupby('City')[['Science','Total']].mean().reset_index()
city_avg_totals.columns = ['City','Avg_Sc_Marks' ,'Avg_Total_Marks']

print("\nAverage total marks by city:")
print(city_avg_totals[city_avg_totals["Avg_Total_Marks"]>250])

print(city_avg_totals[city_avg_totals['Avg_Sc_Marks']==city_avg_totals["Avg_Sc_Marks"].max()]['City'])

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# --- Assuming df already loaded from Part 5 ---
# It contains: Name, City, Maths, Science, English, Total, Rank, Average, Flag

# 1 Distribution of Total Marks
plt.figure(figsize=(7,4))
sns.histplot(df["Total"], kde=True, bins=6, color="skyblue")
plt.title("Distribution of Total Marks")
plt.xlabel("Total Marks")
plt.ylabel("Number of Students")
plt.show()

# 2 Boxplot to detect outliers in marks
plt.figure(figsize=(8,5))
sns.boxplot(data=df[["Maths","Science","English"]], palette="Set2")
plt.title("Subject-wise Marks Distribution (with Outliers)")
plt.ylabel("Marks")
plt.show()

# 3 Compare average marks by subject and city
avg_subject_city = df.groupby("City")[["Maths","Science","English"]].mean().reset_index()

plt.figure(figsize=(8,5))
avg_subject_city.plot(x="City", kind="bar", figsize=(8,5))
plt.title("Average Subject Marks per City")
plt.xlabel("City")
plt.ylabel("Average Marks")
plt.legend(title="Subjects")
plt.show()

# 4 Relationship between Science and Total marks
plt.figure(figsize=(6,5))
sns.scatterplot(x="Science", y="Total", hue="City", style="City", data=df, s=90)
plt.title("Science vs Total Marks")
plt.xlabel("Science Marks")
plt.ylabel("Total Marks")
plt.show()

# 5 Rank vs Total line trend
plt.figure(figsize=(6,4))
sns.lineplot(x="Rank", y="Total", data=df, marker="o")
plt.title("Trend of Total Marks by Rank")
plt.xlabel("Rank")
plt.ylabel("Total Marks")
plt.show()

# 6 Heatmap of city-wise subject correlation
plt.figure(figsize=(7,5))
corr = df[["Maths","Science","English","Total"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Marks")
plt.show()

# 7 Identify top scoring subject overall
subject_means = df[["Maths","Science","English"]].mean()
top_subject = subject_means.idxmax()
print("Top performing subject overall:", top_subject)
print(subject_means)

# 8 Create a city performance summary
city_summary = (
    df.groupby("City")[["Maths","Science","English","Total"]]
      .mean()
      .round(2)
      .sort_values("Total", ascending=False)
)
print("\n City-wise Performance Summary:")
print(city_summary)
```
